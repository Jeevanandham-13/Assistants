```python
import fitz  # PyMuPDF
from docx import Document
from docx.shared import Inches

def convert_pdf_to_word(pdf_path, docx_path):
    # Open the PDF file
    pdf_document = fitz.open(pdf_path)
    
    # Create a new Word document
    doc = Document()
    
    # Iterate through each page in the PDF
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        text = page.get_text("text")
        
        # Add text to the Word document
        if text.strip():  # Add text only if it's not empty
            doc.add_paragraph(text)
        
        # Extract images from the page
        for image_index, img in enumerate(page.get_images(full=True)):
            xref = img[0]
            base_image = pdf_document.extract_image(xref)
            image_bytes = base_image["image"]

            # Save the image temporarily
            image_filename = f"image_page_{page_num+1}_{image_index+1}.png"
            with open(image_filename, "wb") as image_file:
                image_file.write(image_bytes)
            
            # Add the image to the Word document
            doc.add_picture(image_filename, width=Inches(6))

    # Save the Word document
    doc.save(docx_path)
    print(f"PDF successfully converted to Word document: {docx_path}")

# Replace with the path to your PDF file
PDF_PATH = 'D:/POC/automobilereports/src/static/reports/BL24C0224.pdf'

# Path to save the converted Word document
DOCX_PATH = 'report.docx'

convert_pdf_to_word(PDF_PATH, DOCX_PATH)
```

```python
from pdf2docx import Converter

def convert_pdf_to_word(pdf_path, docx_path):
    # Create a Converter object
    cv = Converter(pdf_path)
    
    # Convert PDF to DOCX
    cv.convert(docx_path, start=0, end=None)  # start=0, end=None converts all pages
    
    # Close the Converter
    cv.close()
    print(f"PDF successfully converted to Word document: {docx_path}")

# Replace with the path to your PDF file
PDF_PATH = 'D:/POC/automobilereports/src/static/reports/BL24C0224.pdf'

# Path to save the converted Word document
DOCX_PATH = 'reports.docx'

convert_pdf_to_word(PDF_PATH, DOCX_PATH)
```
