Speech or voice to Text

```python
import speech_recognition as sr

# Initialize the recognizer
recognizer = sr.Recognizer()

# Record audio from the microphone
with sr.Microphone() as source:
    print("Speak something...")
    audio = recognizer.listen(source)

# Recognize speech using Google Speech Recognition
try:
    print("Recognizing...")
    text = recognizer.recognize_google(audio)
    print("You said:", text)
except sr.UnknownValueError:
    print("Sorry, could not understand audio.")
except sr.RequestError as e:
    print("Error fetching results from Google Speech Recognition service:", e)
```

Audio file to Text

```python
import speech_recognition as sr

# Define the audio file path
audio_file = r"C:\Users\JeevanandhamGanesan\Downloads\output_audio.wav"

# Initialize the recognizer
recognizer = sr.Recognizer()

# Load the audio file
with sr.AudioFile(audio_file) as source:
    # Record the audio from the file
    audio_data = recognizer.record(source)

    try:
        # Use Google Speech Recognition to transcribe the audio
        text = recognizer.recognize_google(audio_data)
        print("Transcription: ", text)
    except sr.UnknownValueError:
        print("Google Speech Recognition could not understand the audio.")
    except sr.RequestError as e:
        print("Could not request results from Google Speech Recognition service; {0}".format(e))
```

TEXT TO SPEECH


```python
from gtts import gTTS
from IPython.display import Audio
import io

# Text you want to convert to speech
text = """Transcription:  learn while you play story based skill development games and experiences are based concepts ITC is primary education program is designed to provide children from weaker sections access to education its focus on learning outcomes and detention focus on Universal access and detention of Gender and social category gaps in elementary education and improve the quality of learning operational 13 States the program has benefited in 864000 children till March 2022 so this is the speech thank you
"""

# Specify language (optional)
language = 'en'

accent = 'uk'

# Slow down the speech (optional)
slow_speed = False

# Create a gTTS object
tts = gTTS(text=text, lang=language + '-' + accent, slow=slow_speed)

# Convert the audio to bytes
audio_bytes = io.BytesIO()
tts.write_to_fp(audio_bytes)

# Play the audio directly in the notebook
Audio(data=audio_bytes.getvalue(), autoplay=True)
```
